```python
import numpy as np
import pandas as pd 
```

```python
data = pd.read_csv('../Data/excel.csv', sep = ';', engine = 'python')
```

```python
for i in data.columns:
    print(i, data[i].unique(), data[i].dtype)
```

```python
# colonne superflue
inut = ['campione','livello','orario_ita','orario_mat','_merge1', '_merge2', '_merge3']
#'Sì' scritti male
corr_si = ['freq_asilo_nido', 'freq_scuola_materna', 'D3bis_dir', 'D3bis_ins', 'D4_a','D4_b', 'D4_c', 'D4_d', 'D4_e', 'D4_f', 'D4_g','D6',\
          'D9_a', 'D9_b', 'D9_c', 'D9_d', 'D9_e', 'D9_f','D17','D17bis', 'D20', 'D29', 'D29bis_a','D29bis_b','D29bis_c',\
          'D30_a','D30_b','D30_c','D30_d','D30_e','D30_f','D31']
#parse int
parse_i = ['voto_scritto_ita','voto_orale_ita','voto_scritto_mat','voto_orale_mat',\
           'voto_scritto_ita_cont','voto_orale_ita_cont','voto_scritto_mat_cont']
#Parse codice.
parse_c = ['titolo_padre', 'prof_padre','titolo_madre', 'prof_madre'] 
#parse_float
parse_f = ['Fattore_correzione', 'ita8', 'ESCS', 'school_score_ita', 'school_escs', 'prop_fem', 'prop_imm1', 'prop_imm2','prop_late','pu_ma_no',\
          'part_1_no','part_2_no' ,'part_3_no' ,'part_4_no','proc_1_no', 'proc_2_no','proc_3_no',  'WLE_MAT_200', 'WLE_MAT_200_CORR',\
          'mate8','pu_it_no', 'WLE_ITA_200', 'WLE_ITA_200_CORR', 'ita6', 'mate6', 'school_score_mate']
#nan classes
nan_classes = ['Mancante d','Non dispon','7. Non dis','10. Non di','Non classi','Senza voto','Dato manca']          
```

```python
def replace_with_nan(col):
    for i in nan_classes:
        col = col.replace(i,np.nan)
    return col

def parse_int(col):
    col.fillna(-99999, inplace = True)
    c = col.apply(int)
    c[c == -99999] = np.nan
    return c

def parse_float(col):
    col.fillna('-9,9999', inplace = True)
    c = col.apply(lambda x: x.replace(',','.')).apply(float)
    c[c == -9.9999] = np.nan
    return c

def correct_si(col):
    col.fillna('-9,9999', inplace = True)
    c =  col.apply(lambda x: x.replace('<ec>','i'))
    c[c == '-9,9999'] = np.nan
    return c

def correct_D15(col):
    col.fillna('-9,9999', inplace = True)
    c = col.apply(lambda x: x.replace('C<U+0092><e8> una pr','Molti genitori'))
    c = c.apply(lambda x: x.replace('Le pressio','Pochi genitori')) 
    c[c == '-9,9999'] = np.nan
    return c

def correct_cod(col):
    col.fillna('-999999', inplace = True)
    c = col.apply(lambda x: x[3:]) 
    c[c == '9999'] = np.nan
    return c
```

```python
data = data.drop(inut,axis = 1)
```

```python
for c in data.columns:
    data[c] = replace_with_nan(data[c])
```

```python
for c in parse_f:
    data[c] = parse_float(data[c])
```

```python
for c in corr_si:
    data[c] = correct_si(data[c])
```

```python
for c in parse_c:
    data[c] = correct_cod(data[c])
```

```python
for c in parse_i:
    data[c] = parse_int(data[c])
```

```python
data['D15'] = correct_D15(data['D15'])
data['D15bis'] = correct_D15(data['D15bis'])
```

```python
for i in data.columns:
    print(i, data[i].unique(), data[i].dtype)
```

```python
data.to_csv('../Data/cleaned_invalsi.csv')
```
